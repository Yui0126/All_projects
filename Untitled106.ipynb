```python
from tqdm import tqdm 
pcts = []

print("AGE | pct insulin ")
for i in tqdm(range(18, 80)):
  twenty = df[df['_AGE80'] == i]
  isw    = twenty[twenty['INSULIN'] == 1]
  
  pct = isw.shape[0] / twenty.shape[0] * 100
  #print("%i   | %.2f"%(i, pct))
  pcts.append(pct)


  
```

```python

fig, ax = plt.subplots(1, 1, figsize=(8, 4))
plt.plot(range(18, 80), pcts, '-', alpha=0.87, label="people using insulin ")
plt.xlabel("Age")
plt.yscale("log")
plt.ylabel("Percentage of age group using insulin ")
plt.title("$$$ from novo nordisk")
plt.legend()
plt.tight_layout()
```
